```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("house-prices.csv")
```

```python
df.iloc[:, 1] = df.iloc[:, 1]/1000
df
```

```python
w = 0
b = 0
last = 1
seclast = 0
x = np.array(df.iloc[:,2])
y = np.array(df.iloc[:,1])
while(abs(last-seclast)>0.001):
    djw = 0
    djb = 0
    j = 0
    z = np.sum(y)
    a = np.sum(x)
    c = np.sum(x*x)
    j = (w*a - z + 128*b)**2
    djw = w*c - np.dot(y,x) + b*a
    djb = w*a - z + 128*b
    j = j/(2*128)
    djw = djw/128
    djb = djb/128
    print(j)
    seclast = last
    last = j
    new_w = w - 1e-7*djw
    new_b = b - 1e-7*djb
    w = new_w
    b = new_b
```
